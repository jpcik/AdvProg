# Exercises

### 1. Create a list of numbers which are consecutive multiples of 3: 3, 6, 9, 12, ... 81


### 2. Given a list like this, [3,5,2,3,9,4,2] generate a new list replacing each element by its own triple (e.g., instead of 3 -> 6).


### 3. Given a list of double numbers, find only those whose decimal part is greater than 0.5 (e.g. 45.6 but not 23.1).

### 4. Based on a list like ["jim","lara","linda","john","barry"] Create a key-value Map consisting of each string Capitalized as key, and the value being the number of characters of each. Then filter only those whose key starts by 'L' and show the maximum character length.
