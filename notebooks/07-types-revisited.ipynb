# Scala types revisited

A bit more abou the Scala type systems.

Types can be inferred by the compiler, as in this `Float`:

```scala
val num = 3f
```

All basic value types (e.g., `Int`, `Float`, `Double` inherit from `AnyVal`.

```scala
val any:AnyVal = num
```

```scala
3 + any
```

However `AnyVal` has little functionality, so most methods are applicable in the specific subtypes.

```scala
3 + any.asInstanceOf[Float]
```

String instances inherit from the `AnyRef` superclass. Types can also be inferred in this case.

```scala
val name = "Nelson"
```

```scala
val any:AnyRef = name
```

```scala
any + "Mandela"
```

Again, only the subtypes have most of the specific methods and fucntionalities.

```scala
any.asInstanceOf[String] + " Mandela"
```

 ### Dealing with null values
 
 The `Null` tyes exists in Scala.

```scala
val nada:Null = "something"
```

But it can only take the `null` value, all other values would raise an error.

```scala
var nada:Null = null
```

```scala
nada = "something"
```

```scala
val niente:Nothing = null
```

```scala
val niente:Nothing = throw new Exception
```

### Product type

Arbitrary classes can be seen as instances of a `Product` type. 

```scala
case class Student(name:String, age:Int)
```

```scala
val p:Product = new Student("Pete",23)
```

With a `Product` it is possible to iterate over the attributes as collections.

```scala
for elemName <- p.productElementNames do
    println(elemName)
```

```scala
for element <- p.productIterator do
    println(element.getClass)
```

```scala
p.productIterator
```

```scala
p.productIterator.foreach(a=>println(a))
```

### Tuples

Tuples are convenient as type-safe data containers.

```scala
val tuple = ("Jim",5.5,6,3.4)
```

It is possible to access the `Tuple` elements with special access methods by order:

```scala
tuple._1
```

```scala
tuple._3
```

```scala
tuple._5
```

```scala
tuple.size
```

It is also possible to declare multiple values within a tuple construct:

```scala
val (name,age,grade) = ("Joe",34,6)
```

```scala
grade
```

Tuples ar also handy for `match` `case` selection, for example using wildcards:

```scala
tuple match
    case ("Jim",_,_,_) => println("found")
    case _ => println("unfound")
```

```scala
tuple match
    case (s:Int,_,_,_) => println("gotcha")
    case _ => println("nein")
```

Tuples can also be concatenated.

```scala
val tuple2 = ("Nancy",5.5) ++ (4.5,5.3)
```

```scala
tuple2._3
```

```scala
tuple2(2)
```

And converted into lists.

```scala
tuple2.toList
```
