# Scala: Strings

Strings and chars. Types can be inferred by the compiler.

```scala
val course = "History"
```

```scala
val grade='B'
```

## Strings and AnyRef

Strings, as other complex types in Scala, inherit from `AnyRef`.

```scala
val someRef:AnyRef = "my string"
```

```scala
someRef + "adding"
```

As we can see, the methods for manipulating the strings are not available in AnyRef.

We can cast any `String` as `AnyRef`, as a consequence of that inheritance.

```scala
val oneString:String = "one string"

val oneRef:AnyRef = oneString
```

## String interpolation

String interpolation can be useful to mix strings and variables which are preceded by the `$` symbol.

```scala
val note = s"Your grade in $course is $grade."
println(note)
```

String interpolation can also be ised to introduce full expressions enclosed in `${ }`

```scala
val grade1 = 5.5
val grade2 = 3.2
def add(x:Int)=x+2
s"Your final grade is ${(grade1 + grade2) /2}. Bye."
```

## Other common String features and operations

Multi-line strings can be specified using triple `"""  """`

```scala

val poem = """The grass is green
            The sky is blue
            The river flows
            And the trees sleep."""
```

Common methods to split, trim, substring, etc. are available.

```scala
"sometext".split('e')
```

```scala
"231,;231,213;324234".split(",|;")
```
