# Anonymous functions

### Identity function and apply method

Identity function. Doesn't do much.

```scala
(x:Any) => x
```

We can apply to anything basically:

```scala
((x:Any) => x) (4)
```

The parentheses `()` for the application of the function is really an `apply` method.

```scala
((x:Any) => x).apply(4)
```

what if we apply it to more parameters?

```scala
((x:Any) => x) (4) (5)
```

We have a type error, the anonymous function has only one parameter, so it is a type error.

Otherwise this function can tyke any type:

```scala
((x:Any) => x) ("Hallo")

((x:Any) => x) (34.5)

((x:Any) => x) (List("Hallo",34,4.5))
```

### Function application

The identity function is not that interesting, we can try others, like a multiplication:

```scala
(x:Int) => x*x 
```

We can apply this function to an expression or a value. The following code shows how we can apply the function to the value 5. This means that the `x` variable takes thes value 5.

```scala
((x:Int) => x*x )(5)
```

Here the function had the type implicit. We can specify it explicitly:

```scala
((x:Int) => x*x ): (Int => Int)
```

We can modify the type, for instance to indicate that the resturn type is a `Double`

```scala
((x:Int) => x*x ): (Int => Double)
```

To use this function more easily we can reference it with a `val`. 

```scala
val triple = (x:Int) => x*3
```

```scala
triple(5)
```

We can define even more complex functions. In this case it is a composition of two functions. The notation is a bit complex. 

```scala
( (x:Int) => ( (y:Int) => y+x) )(3) (5)
```

```scala
val f = ( (x:Int) => ( (y:Int) => y+x) )(3) 
```

```scala
f (5)
```

Nevertheless we can simplify the notation a bit, so that the function can be set to receive two different parameters. 

```scala
((x:Int,y:Int) => x+y )(3,5)
```

### Returning functions


In this function the return is another function that takes a `Double` and returns another `Double`:

```scala
((x: Int) => ( (y:Double) => (x+y)) )
```

We can apply this function to a value that will replace the variable `x`:

```scala
((x: Int) => ( (y:Double) => (x+y)) ) (2)
```

For convenience we can give this funciton aname through a value `f`, and then use the `apply` method, or `()`:

```scala
val f = ((x: Int) => ( (y:Double) => (x+y)) )
f (2)
```

As the return value of `f`, is another function, we can simply apply it again to another value, or give it yet another name `g`.

```scala
(f(2))(3)
```

```scala
val g = f(2)
g(8)
```

```scala
g(3)
```

### Functions as parameters

Just as we can return function, we can pass them asparameters of other functions. Consider the following function, it takes a parameter `f` that takes an `Int` and returns an `Int`:

```scala
val myFun = ((f:Int=>Int) => f(4)) 
```

To apply this function we will need to pass some function with the `Int => Int` signature:

```scala
myFun( ((x:Int) => x+x)  )
```

We could apply it to other functions with the same signature:

```scala
val multiply = ((x:Int) => x*x)
val triple = ((x:Int) => 3*x)
```

```scala
myFun(multiply)
```

```scala
myFun(triple)
```

These functions as parameters are used a lot in collection combinators. If we take another function:

It can be used as the parameter of a map function. We will see this a lot in collections.

```scala
(1 to 10).map((x:Int) => x+5)
```

```scala
(1 to 10).map(x => x+5)
```

```scala
(1 to 10).map(_+5)
```
