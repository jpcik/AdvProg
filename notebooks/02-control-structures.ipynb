# Control Structures

## if expressions

This control structures works as in other languages: `if` condition `then` expression `else if` condition `then` expression:

```scala
val age = 34

if age < 28 then 
  println("you are young")
else if age <38 then
  println("your are not that old")
else
  println("age does not matter")
```

```scala
val age = 34

if age < 28 then     
    "you are young"
else if age <38 then 
    "your are not that old"
else                 
    "age does not matter"
```

One-liner ifs are possible following this syntax:

```scala
val grade = 3.1

val decision = if grade >= 4 then "pass" else "fail"

println(decision)
```

## for loops

For loops allow iterating over collections, lists arrays, etc. 

```scala
val nums = List(1,2,3,4,5)

for i <- nums do println(i)
```

Conditions can be inserted in the `for` declaration. Here we iterate over only pair numbers.

```scala
for i <- nums if i % 2 == 0
do println(i)
```

We can have multiple iteration variables in a loop.

```scala
for 
  i <- 1 to 3 
  j <- 1 to 5
do 
  println(s"i:$i -> j:$j")
```

```scala
for i <- nums 
do 
  val calc = i * 4/(10-i)
  println(calc)
```

```scala

val result = 
  for i <- nums 
  do 
    val calc = i * 4/(10-i)
    println(calc)
result
```

```scala

val result = 
  for i <- nums 
  yield 
    val calc = i * 4/(10-i)
    calc
```

## Match expressions

Match expressions allow executing different code segments depending on a selection criteria.

```scala
val day=3

day match 
  case 0 => print("sunday")
  case 1 => print("monday")
  case 2 => print("tuesday")
  case 3 => print("wednesday")
  case 4 => print("thrusday")

```

```scala
import annotation.switch
val day=3

(day: @switch) match 
  case 0 => print("sunday")
  case 1 => print("monday")
  case 2 => print("tuesday")
  case 3 => print("wednesday")
  case 4 => print("thrusday")

```

```scala
val grade = 'B'

grade match
  case 'A' => println("you are excellent")
  case 'B' => println("you are good")
  case 'C' => println("you are fine")
  case _   => println("you failed")
  
```

```scala
val swissGrade = grade match
  case 'A' => 6
  case 'B' => 5
  case 'C' => 4
  case _   => 3

println(swissGrade)
```

```scala
val grade = 'C'

val swissGrade = grade match
  case 'A' => 6
  case 'B' => 5
  case 'C'|'D' => 4
  case _   => 3

println(swissGrade)
```

We will see much more complex match expressions later with complex types.

```scala
val something = "2"
something match
  case "1" => "ok"
  case 2   => "bad"
  case 3.5 => "bad"
  case _   => "maybe"

```

```scala
val something:Any = 2
something match
  case s:String => s"we got string $s"
  case i:Int    => s"we got int ${i+0}"
  case f:Float  => s"we got float $f"
  case _        => "got something else"
```
