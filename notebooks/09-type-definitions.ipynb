### Type aliases and Type members

Sometimes it is useful to decalre type aliases for complex type signatures. 

The `type` keyword is used:

```scala
type Grade=(Int,String)
```

Then we can use the type alias in the code:

```scala
case class GradeSheet(name:String,grades:Seq[Grade])
```

```scala
val jimGrades = GradeSheet("jim",Seq((5,"math"),
                                     (4,"physics")))
```

However this is not replacing the original type which is a tuple, so this code will not work:

```scala
val mathGrade = Grade(3,"history")
```

We can declare othe nested aliases as in this example:

```scala
type Word = String
type Sentence = List[Word]
type Paragraph = List[Sentence]
```

```scala
val p1:Paragraph = List(
                     List("We","code"),
                     List("We","debug"))
```

### Type members

Types are not only aliases, but can be used to define abstract type members. Consider this trait:

```scala
trait Reptile :
    type T
    def feed(food:T):Unit
```

Using this we can declare a class that inherits from `Reptile`, and we also define what `T` is in this case:

```scala
case class Vegetable(name:String)
```

```scala
class Turtle extends Reptile:
    type T = Vegetable
    def feed(food:T) = println( s"yammi $food" )
```

```scala
object carrot extends Vegetable("carrot")
```

```scala
val t = new Turtle
t.feed(carrot)
```
