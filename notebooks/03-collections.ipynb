# Scala: Collections

Arrays in scala can include elements of different types. Generic types exist in Scala. Types can be inferred by the compiler.

```scala
val array1 = Array(4,6,7,3)
```

Common methods are avaiable and accessing individual elements is possible through the apply function using `(i)`

```scala
array1.length
```

```scala
array1(2)
```

```scala
array1.head
```

```scala
array1.tail
```

Arrays are mutable, so the elements can be replaced. Multidimensional arrays can be specified.

```scala
array1(0)=9
array1
```

```scala
val array2 = Array( Array(3,4,6) , Array(2,5), Array(2,7,9) )
array2(0)(1)
```

Lists are immutable. However, the mutable version exist although they are not commonly used.

```scala
val list1 = List(3,4,2)

list1.head
list1.tail
```

Ranges can be specified with boundaries and steps. Lists ans sequences can be generated from these ranges.

```scala
val list2 = (1 to 10)
val list3 = (1 to 10 by 3)
val list4 = (1 until 5)
```

Lists, and most sequences are well equipped with basic methods for common manipulation operations.

```scala
val a = List(10, 20, 30, 40, 10)      // List(10, 20, 30, 40, 10)
```

```scala
a.drop(2)                             // List(30, 40, 10)
```

```scala
a.dropWhile(_ < 25)                   // List(30, 40, 10)
```

```scala
a.filter(_ < 25)                      // List(10, 20, 10)

a.filter(x => x < 25)                      // List(10, 20, 10)
```

```scala
a.slice(2,4)                          // List(30, 40)
```

```scala
a.tail                                // List(20, 30, 40, 10)
```

```scala
a.take(3)                             // List(10, 20, 30)
```

```scala
a.takeWhile(_ < 30)                   // List(10, 20)
```

```scala
// flatten
val a = List(List(1,2), List(3,4))
a.flatten                             // List(1, 2, 3, 4)
```

```scala
// map, flatMap
val nums = List("one", "two")
nums.map(_.toUpperCase)               // List("ONE", "TWO")
nums.flatMap(_.toUpperCase) 
```
