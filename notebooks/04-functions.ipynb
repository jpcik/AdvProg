# Scala: Functions and methods

Methods are indicated with the `def` keyword. A simple method may have no arguments. The body of the method could be a simple value. In this example it is a constant.

```scala
def fun = 3
```

Methods can be directly invoked using the method name. In this case there are no arguments and it does not require parentheses. 

```scala
fun
```

The previous method declaration does not indicate the return type. The compiler can infer the return type, wich was `Int`. This can also be declared explicitly. The return value can be also explicitly indicated with the `return` keyword. 

```scala
def fun(): Int =
  return 4
```

```scala
fun()
```

We can define yet another method that takes an `Int` and returns an `Int`

```scala
def triple(x:Int) = x*3

triple(4)
```

We can redefine it as a function, in this case an anonymous function. This reminds of the lambda notation.

```scala
(x:Int) => x*3 
```

We can apply this function to an expression or a value. The following code shows how we can apply the function to the value 5. This means that the `x` variable takes thes value 5.

```scala
((x:Int) => x*3 )(5)
```

To use this function more easily we can reference it with a `val`. 

```scala
val triple = (x:Int) => x*3
```

```scala
triple(5)
```

We can define even more complex functions. In this case it is a composition of two functions. The notation is a bit complex. 

```scala
( (x:Int) => ( (y:Int) => y+x) )(3) (5)
```

Nevertheless we can simplify the notation a bit, so that the function can be set to receive two different parameters. 

```scala
((x:Int,y:Int) => x+y )(3,5)
```

We will see later that these functions (anonymous or not) can be really useful for composition and for list operations. Otherwise we can transform it into a method, as we saw earlier:

```scala
def adding(x: Int, y: Int) = 
  x+y
```

Methods can of course have more complex bodies, and also potentially throw exceptions.

```scala
def computeArea(width:Int, height:Int) =
  if width >= 0 && height >=0 then
    width*height
  else 
    throw new IllegalArgumentException("must be positive")
```

```scala
computeArea(-3,6)
```

It is also possible to indicate default arguments in a method. 

```scala
def computeGrade(math:Int, history:Int, geology:Int=0) = 
  (math + history + geology) / 3

computeGrade(3,4)
```

Named arguments allow certain liberty and readability. 

```scala
computeGrade(history=4,math=3,geology=3)
```
