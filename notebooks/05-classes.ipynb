# Scala: Classes and Traits

The simplest way of creating a user is simply by gicing it a name. This class has no attributes.
But we can instantiate it with the `new` keyword.

```scala
class User

val joe = new User
```

A more interesting class may have some attributes. By declaring them, we also declare the default constructor of the class. But still this class has no body.

```scala
class User(val name:String, age:Int)

val zoe = new User("Zoe",34)
```

We can try to access the attributes, but you'll notice that it will fail at compile time. This is because these attirbutes are actually not available outside the class. 

```scala
zoe.name
```

```scala
zoe.age
```

To make them available we need to use the `val` keyword. And they will be immutable as you may imagine. Now we can access them after we create an instance.

```scala
class User(val name:String, val age:Int)

val zoe = new User("Zoe",34)
zoe.name
```

But we cannot overwrite the values as they are immutable. This is different from the usual Java Beans, and the getter setter methods.

```scala
zoe.name = "Zoey"
```

## Classes for modeling

```scala
import java.time.LocalDate

class Hotel(val name:String,
            val stars:Int,
            val city:String,rooms:Int,parking:Int=0) 
```

```scala
class Booking(val user:User, val hotel:Hotel,val date:LocalDate) 
```

```scala
class HotelManager(hotel:Hotel) :
  def isBookingOk(booking:Booking) = 
    (hotel.name==booking.hotel.name) &&
    (booking.date.isAfter(LocalDate.now))
```

```scala
val palace = new Hotel("Palace",5,"Lausanne",50)
val paix = new Hotel("Hôtel de la Paix",4,"Lausanne",30,20)
```

```scala
val palaceManager = new HotelManager(palace)
```

```scala

val booking = new Booking(zoe,palace,LocalDate.parse("2023-10-01"))

palaceManager.isBookingOk(booking)
```

## Using traits

```scala
 class Song(val title:String)
```

```scala
trait Singer:
  def sing(song:Song):Unit
 
trait Songwriter:
  def compose(title:String):Song
  
```

```scala
class SingerSongwriter extends Singer with Songwriter:
  def compose(title:String) =
    new Song(title)

  def sing(song:Song) =
    println(s"♫♫  ${song.title} ♫♫")
```

```scala
trait ClassicSinger extends Singer:
  def sing(song:Song) = println(s"lalalala ${song.title}")
```

```scala
class OperaSinger extends ClassicSinger:
  override def sing(song:Song) =  println(s"lololo ${song.title}")
```

```scala
val placido=new OperaSinger
placido.sing(new Song("Figaro"))
```

## A bit of objects and implicit magic

```scala
class Poem(val title:String)

object Songs :
    def fromPoem(p:Poem):Song = new Song("Song: "+p.title)
```

```scala

val iliad= new Poem("Iliad")
placido.sing(Songs.fromPoem(iliad))

```
