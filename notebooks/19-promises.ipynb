## Promises

A promise can be thought of as a writable, single-assignment container, which completes a future.

```scala
import scala.concurrent.Promise
import scala.concurrent.Future
import concurrent.ExecutionContext.Implicits.global
```

A promise `p` completes the future returned by `p.future`. This future is specific to the promise `p`.

```scala
val p = Promise[Int]
```

```scala
val f = p.future
```

A promise can be used to successfully complete a future with a value (by completing the promise) using the `success` method. 

```scala
def getTemperature = 
    Thread.sleep(2000)
    37
```

## Summarizing Futures

 - Construct futures to run tasks off of the main thread

 - Intended for one-shot, potentially long-running concurrent tasks
 
 - A future starts running as soon as you construct it
 
 - work with for expressions, and come with a variety of callback methods 
 
 - simplify the process of working with concurrent threads

 - don’t have to concern with the low-level details of thread management
 
 - handle the result of a future with callback methods or transformation methods
