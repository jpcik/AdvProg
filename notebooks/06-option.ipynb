# Scala: Option

Before intorducing options let's create a Map of grades.

```scala
val grades = Map("math"->3,"biology"->5,"history"->3)
```

We can apply all sorts of operations in a Map.

```scala
grades.values.sum/grades.size.toFloat
```

But how about a course in which we have no grade yet. One may be tempted to use `null`

```scala
val grades = Map("math"->3,"biology"->5,"history"->3,"physics"->null)
```

But then things go wrong if we want to do the calculations again. The null has messed up the data type.

```scala
grades.values.sum/grades.size.toFloat
```

With Option we can create an object that can be either `None` or `Some`. If it is `Some` it han hold the value that we want.

```scala
val grade:Option[Int]= Some(4)
```

With an option, we can always verify if it has a value or not:

```scala
grade.isEmpty
```

```scala
grade.isDefined
```

And we can get the value if needed:

```scala
grade.get
```

Now back to the grades example, le'ts do it with Option:

```scala
val grades = Map("math"->Some(3),"biology"->Some(5),"history"->Some(3),"physics"->None)
```

```scala
grades.values.filter(g=>g.isDefined)
```

Now we can do the computation as well. Two different ways to do it. Notice that with the flatten we get rid of the Option. So this eliminates the None values.

```scala
grades.values.filter(g=>g.isDefined).map(g=>g.get).sum/grades.size.toFloat
```

```scala
grades.values.flatten.sum/grades.size.toFloat
```
