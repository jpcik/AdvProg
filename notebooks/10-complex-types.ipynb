# Complex types

### Intersection types

Uses the `&` to define the type intersection

```scala
trait Singer:
    def sing(song:String):String

trait Songwriter:
    def write(lyrics:String):String
```

```scala
def recordSong(musician:Singer & Songwriter)=
    val song = musician.write("lala")
    musician.sing(song)
```

Objects must inherit (or be) of the both types of the intersection

```scala
object mike extends Singer, Songwriter {
    def sing(song:String) = s"$song"
    def write(lyrics:String) = s"$lyrics"
} 

recordSong(mike)
```

We can also use it in Generics, as in Lists. Elements must comply to the complex type.

```scala
val weirdList = Array[Int & Double](3,4,5)
```

Intersection types can also be defined in aliases for convenience.

```scala
type Musician = Singer & Songwriter
```

Some sort of Duck typing is also possible thanks to intersection types.

```scala
trait Human:
    def speak = "hallo"
trait Bat:
    def speak = "iiii"
trait Rat:
    def speak = "skweek"

def speakNow(individual: Human & Bat)=
    individual.speak
```

```scala
object batman extends Human, Bat:
    override def speak = super.speak+" i am batman"
```

```scala
batman.speak
```

```scala
speakNow(batman)
```

Finally, we can see that extending classes with `with` results also in intersection types.

```scala
trait T1
trait T2
class C
class CC extends C, T1, T2
val cc = new C with T1 with T2
```

### Union Types

```scala
type Grade = Int | String

val grades = Seq[Grade]
```

```scala
val grades = Seq[Grade](3,5,"excellent")
```

```scala
val grades = Seq[Grade](3,5,"excellent",4.5)
```

```scala
val moreGrades = List(3,5,6,"really bad",3.2)
```

```scala
moreGrades.foreach(g=>g match {
    case g:Grade => println("fine")
    case _ => println("not fine")
})
```

```scala
def evaluate(grade:Int) : Grade =
    if grade == 6 then "Cum Laude"
    else grade
```

```scala
evaluate(5)
```

```scala
evaluate(6)
```

```scala
val grades:Seq[Grade] = Seq(2,5,"A",6,"B")
```

```scala
def qualify(g:Grade) = g match
    case i:Int => s"numeric grade is $i"
    case s:String => s"literal grade is $s"
```

```scala
grades.map(qualify) 
```

### Structural Types

```scala
type Dancer = {
    def dance:String
}

```

```scala
val jenny:Dancer = new :
    def dance = "dancing now"
```

```scala
jenny.dance
```

```scala
import reflect.Selectable.reflectiveSelectable
jenny.dance
```

```scala
val doc:Dancer = new {
    def dance = "another dance"
}
```

```scala
doc.dance
```

```scala
case class Waltzer(name:String):
  def dance = s"$name dancing the waltz!"
```

```scala
val schumann:Dancer = Waltzer("schumann")
schumann.dance
```

```scala
class Ballerina:
    def danceBallet = "ballet ballet"
```

```scala
val pavlova:Dancer = new Ballerina
```

```scala
val pavlova:Dancer = new Ballerina {def dance = danceBallet}
```

```scala
pavlova.dance
```
