# Scala: types and expressions

Every expression is a function that results in a value.

## Basic data types

All types in scala inherit from `Any`, including basic data types as Int or Boolean.

```scala
3
```

```scala
true
```

Expressions result in a value, as for instance in arithmetic expressions:

```scala
1+3
```

```scala
2.5+4
```

```scala
math.Pi
```

```scala
math.log(1787)
```

Strings are compatible with the Java `String` objects. 

```scala
"somestring"
```

```scala
"somestring"+4
```

Types are not always required to be stated explicitly. When needed, the compiler will infer the data type.

```scala
2*6.0/7
```

Expressions can sometimes lead to runtime exceptions.

```scala
3/0
```

## Type casting

Types may not always be explicitly declared. The following `Int` value type is inferred by the compiler. Casting from one type to another is sometimes possible, e.g. `Int` to `Float`.

```scala
val a = 4
val b:Float = a 
```

```scala
val x: Long = 987654321
val y: Float = x 
```

`Char` expressions can be cast to `Int`

```scala
val face: Char = '☺'
val number: Int = face  
```

Casting is not always possible and may lead to compilation errors.

```scala
val x: Long = 987654321
val y: Float = x  
val z: Long = y  
```

## Values and variables

`val` declarations can be used to reference an expression, e.g. a value or even a function. For instance it can reference a immutable value:

```scala
val x = 23
```

```scala
x
```

```scala
x + x
```

```scala
val name = "Catbert"
```

The type of the value or expression can be inferred or stated explicitly. If the Type declaration and the expression type do not match, compilation errors can be thrown.

```scala
val num:Double = 4
```

```scala
val mix = "text" + 23 + true
```

```scala
val i:Int = 3.5
```

## Mutable vs Immutable

Scala does allow mutable variables using `var`. However, this is mostly discouraged in order to prevent side effects and realted issues. Using `val` forces the programmer to stick to immutability.

```scala
val aNumber = 4
```

```scala
aNumber = 6
```

```scala
var anotherNumber = 5
```

```scala
anotherNumber = 3
anotherNumber
```

Although `null` is available, it's not generally used, and other alternatives (e.g. Options) are preferred.

```scala
var x = null
```

```scala
x = 4
```
