Every expression is a function that results in a value.

## Everything has a type

Expressions result in a value, which has a defined type. For instance this number:

```scala
4
```

It is of type `Int`. The compiler knows the type which is evidently inferred as it is an integer number extracted from the parser.

The same can be applied to a double value:

```scala
4.5
```

Or for a boolean value:

```scala
true
```

## Expressions have types

Expressions result in a value, as for instance in arithmetic expressions:

```scala
1+3
```

The expression results also has a type, in this case `Int`.

For the same operator (`+`) the resut type can be different, depending on the operands:

```scala
2.5+4
```

The same goes for other expressions, e.g., other mathematical operations:

```scala
math.Pi
```

```scala
math.log(1787)
```

```scala
2*6.0/7
```

Strings also have a type, and expressions can comine strings with other types.

Strings are compatible with the Java `String` objects. 

```scala
"somestring"
```

```scala
"somestring"+4
```

## Values and variables

`val` declarations can be used to reference an expression, e.g. a value or even a function. For instance it can reference a immutable value:

```scala
val x = 23
```

Types are not always required to be stated explicitly. When needed, the compiler will infer the data type.

In this case the type of `x` was inferred by the compiler. It is `Int`.

```scala
x
```

```scala
x + x
```

In this case the compiler understands that it is of type `String`:

```scala
val name:String= "Catbert"
```

```scala
val mix = "text" + 23 + true
```

## Explicit and implicit type declaration

The type of the value or expression can be inferred or stated explicitly. If the Type declaration and the expression type do not match, compilation errors can be thrown.

In this example we **explicitly** say that `num` is of type `Double`:

```scala
val num:Double = 4
```

In this other example `i` is declared of type `Int`, but the assigned value is not an integer, so we have a problem.

```scala
val i:Int = 3.5
```

This is a compilation error, so in Scala you do not have to wait until runtime to find these problems. It is a statically and strongly typed language.

## Basic data types

All types in scala inherit from `Any`, including basic data types as Int or Boolean.

```scala
val someValue:Any = 14
```

However it is not super useful per se, as we do not know what kind of value it is:

```scala
someValue + 10
```

`Any` has two main subclasses. `AnyVal` represents values, like integers, booleans, etc.

```scala
val anotherValue:AnyVal = 24
```

Still, this type is too high level to do things like arithmetic operations, as we do not know if its is a number.

```scala
anotherValue + 10
```

If we want to really do numerical operators then we can explicitly declare using, for example a `Short` number:

```scala
val shortValue:Short = 34
```

```scala
shortValue + 10
```

Expressions can sometimes lead to runtime exceptions.

```scala
3/0
```

## Type casting

Types may not always be explicitly declared. The following `Int` value type is inferred by the compiler. Casting from one type to another is sometimes possible, e.g. `Int` to `Float`.

```scala
val a = 4
val b:Float = a 
```

```scala
val x: Long = 987654321
val y: Float = x 
```

`Char` expressions can be cast to `Int`

```scala
val face: Char = '☺'
val number: Int = face  
```

Casting is not always possible and may lead to compilation errors.

```scala
val x: Long = 987654321
val y: Float = x  
val z: Long = y  
```

## Mutable vs Immutable

Scala does allow mutable variables using `var`. However, this is mostly discouraged in order to prevent side effects and realted issues. Using `val` forces the programmer to stick to immutability.

```scala
val aNumber = 4
```

```scala
aNumber = 6
```

```scala
var anotherNumber = 5
```

```scala
anotherNumber = 3
anotherNumber
```

Although `null` is available, it's not generally used, and other alternatives (e.g. Options) are preferred.

```scala
var x = null
```

```scala
x = 4
```

This is invalid, as an `Int` value has a different type than null. It is generally preferable **not to use** null values in Scala code.

We can also use suffixes to help the compiler know we are dealing with specific types, like floats:

```scala
val count = 2.0f;
val cost = 1.3f;
val result = count - cost;
```
