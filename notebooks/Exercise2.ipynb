# Exercise 2: Types in Scala

Consider a hospital with different units: e.g. Oftalmology, Paediatrics, Urology, Surgery, Psychiatry, Endocrinology, etc.
There are also multiple doctors, specialized in these and other areas. 
We can also have patients with health conditions and who are visiting these uits and making consultations with one or more of these doctors.

 - 1. Create a class model in Scala of the hospital, including these uints, doctors, patients, consultations, and other elements you consider important. Mandatory use of inheritance, traits, and case classes.
 - 2. Create methods relative to these classes, e.g. for creating a consultation, reserve an appointment, etc. Mandatory use of typed parameters in either classes or methods.
 - 3. Include one use of Type Restrictions and explain why it is useful in this context.
 - 4. Include at elast one example of Covariance and Contravariance, and explain why it is useful in your use example.
